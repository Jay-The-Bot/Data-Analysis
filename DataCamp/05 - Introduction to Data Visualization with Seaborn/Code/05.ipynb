```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
# The box plot here shows the distribution of total bill paid per table across different days of the week.
# From this plot we can easily see that median bill is higher on Saturday and Sunday, but the spread of distribution is also larger.
# This comparison would be much harder to do with other types of visualizations.
g = sns.catplot(x='time', y='total_bill', data=tips, kind='box')
# While seaborn does have "boxplot()" function, we'll be using the "catplot()" function
# We will put the categorical variable "time" on the x-axis and the quantitative variable "total_bill" on the y-axis. 
plt.show()
```

```python
g = sns.catplot(x='time', y='total_bill', data=tips, kind='box', order=['Dinner', 'Lunch'])
plt.show()
```

```python
# Occasionally we may want to omit the outliers from the box plot.
# You can do this using the "sym" parameter
g = sns.catplot(x='time', y='total_bill', data=tips, kind='box', sym='')
plt.show()
# If you pass a empty string into "sym" parameter it will omit the outliers from the your plot altogether.
# Sym can also be used to change the appearance of the outliers instead of omitting them.
# By default, the whiskers extend to 1 point 5 times the interquartile range, or "IQR".
```

```python
# If you want to change the way the whiskers in your box plot are defined, you can do this using the "whis" parameter.
# There are several options for chaining whiskers.
# You can change the range of the whiskers from 1.5 times the IQR (which is the default) to 2 times the IQR by setting the "whis" parameter to (2.0).
# Alternatively, you can have the whiskers define specific lower and upper percentiles by passing in a list of the lower and upper values.
# In this example passing in "[5, 95]" will result in the lower whisker being drawn at the 5th percentile and the upper whisker being drawn at the 95th percentile.
# Finally, you may want to draw the whiskers at the min and max values. You can do it by specifying the lower percentile as 0 and the upper percentile as 100.
g = sns.catplot(x='time', y='total_bill', data=tips, kind='box', whis=[0, 100])
plt.show()
# Note that there are no outliers, because the box and the whiskers cover the entire range of the data.  
```
