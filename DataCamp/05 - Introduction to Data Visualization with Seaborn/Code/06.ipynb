```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
# This point plot uses the tips dataset and shows the average bill among smokers v/s non-smokers.
# The vertical bars extending above and below the mean represent the 95% confidence intervals for that mean.
# Just like the confidence intervals we saw in line plots and bar plots, these confidence levels show us the level of uncertainty we have about these mean estimates.
# Assuming our data is random sample of some population, we can be 95% sure that the true population mean in each group lies within the confidence interval shown.
sns.catplot(x='smoker', y='total_bill', data=tips, kind='point')
plt.show() 
```

```python
sns.catplot(x='smoker', y='total_bill', data=tips, kind='point', estimator=median)
plt.show() 
```

```python
sns.catplot(x='smoker', y='total_bill', data=tips, kind='point', capsize=0.2)
plt.show() 
```

```python
sns.catplot(x='smoker', y='total_bill', data=tips, kind='point', ci=None)
plt.show() 
```

```python
sns.catplot(x='smoker', y='total_bill', data=tips, kind='point', join=False)
plt.show() 
```
